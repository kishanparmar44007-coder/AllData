```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# ------------------------------

# 1. Create Dataset


```python
import numpy as np
import pandas as pd

# Set seed for reproducibility
np.random.seed(42)

n = 150

df = pd.DataFrame({
    "Household_ID": [f"H{i+1}" for i in range(n)],
    "Age_of_Household_Head": np.random.randint(25, 70, n),
    "Household_Income": np.random.randint(10000, 50000, n),
    "Education_Level": np.random.choice(["Primary", "Secondary", "Graduate", "Post-Graduate"], n),
    "Family_Size": np.random.randint(2, 8, n),
    "Owns_House": np.random.choice(["Yes", "No"], n),
    "Urban_Rural": np.random.choice(["Urban", "Rural"], n)
})

df.to_csv("household_data.csv", index=False)

print(df.head())
```

# ------------------------------
# 2. Data Types

```python
print("Data Type\n",df.dtypes)
```

# ------------------------------
# 3. Central Tendency


```python
print("--- Central Tendency ---")

print("Mean",df["Household_Income"].mean())

print("Median",df["Household_Income"].median())

print("Mode",df["Household_Income"].mode()[0])
```

# ------------------------------
# 4. Dispersion

```python
print("--- Dispersion ---")

print("Range",df["Household_Income"].max() - df["Household_Income"].min())

print("Variance",df["Household_Income"].var())

print("Std Dev",df["Household_Income"].std())
```

# ------------------------------
# Skewness & Kurtosis

```python
print("Skewness:", df["Household_Income"].skew())

print("Kurtosis:", df["Household_Income"].kurt())
```

# ------------------------------
# IQR

```python
Q1 = df["Household_Income"].quantile(0.25)
Q3 = df["Household_Income"].quantile(0.75)

IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

# ------------------------------
# 5. Visualizations


# Distribution

```python
plt.Figure()
plt.hist(df["Household_Income"],color='skyblue', edgecolor='black')

plt.title("Distribution of Household Income")
plt.xlabel("Household Income")
plt.ylabel("Frequency")

plt.show()
```

# KDE Plot (Smooth Curve)

# 🔹 Gaussian Distribution

A symmetrical, bell-shaped distribution where mean, median, and mode are equal.


```python
plt.Figure()
sns.kdeplot(df["Household_Income"],fill="true",color="green")

plt.title("Density Curve of Household Income")
plt.xlabel("Household Income")
plt.ylabel("Density")

plt.show()
```

# Boxplot

```python
plt.figure()
sns.boxplot(x="Education_Level", y="Family_Size", data=df, palette='Set2')

plt.title("Family Size by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Family Size")

plt.show()
```

# Scatter Plot

```python
plt.figure()
plt.scatter(df["Age_of_Household_Head"], df["Household_Income"], color='red')

plt.title("Age vs Household Income")
plt.xlabel("Age of Household Head")
plt.ylabel("Household Income")

plt.show()
```

# Skewness Chart

```python
plt.figure()
sns.histplot(df["Household_Income"], kde=True, color='blue')

plt.title("Skewness of Household Income Distribution")
plt.xlabel("Household Income")
plt.ylabel("Frequency")

plt.grid(True)
plt.show()
```

# Kurtosis Chart

```python
plt.figure()
sns.histplot(df["Household_Income"], kde=True, color='green')

plt.title("Distribution Showing Skewness and Kurtosis")
plt.xlabel("Household Income")
plt.ylabel("Frequency")

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))

# Scatter Plot
plt.scatter(df['Age_of_Household_Head'], df['Household_Income'], alpha=0.7)

# Labels
plt.title('Distribution of Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')

# Grid
plt.grid(True)

# Show Plot
plt.show()
```
