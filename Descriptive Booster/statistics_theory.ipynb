
# 📊 Income Distribution & Statistics Theory

---

## 🔹 1. Types of Data

### Numerical Data
Data represented in numerical form on which mathematical operations can be performed.

### Categorical Data
Data represented in the form of categories or labels and cannot be used for mathematical calculations.

---

## 🔹 2. Types of Statistics

### Descriptive Statistics
Used to summarize, organize, and present data.

### Inferential Statistics
Used to make predictions or draw conclusions about a population based on sample data.

---

## 🔹 3. Descriptive Statistics
Descriptive statistics involves methods for organizing, summarizing, and presenting data using numerical measures and graphical tools.

---

## 🔹 4. Mean, Median, Mode

### Mean
\[
\bar{x} = \frac{\sum x}{n}
\]

### Median
The middle value of an ordered dataset.

### Mode
The value that occurs most frequently in the dataset.

---

## 🔹 5. Range, Variance, Standard Deviation

### Range
Difference between the maximum and minimum values.

### Variance
\[
\sigma^2 = \frac{\sum (x - \bar{x})^2}{n}
\]

### Standard Deviation
\[
\sigma = \sqrt{\sigma^2}
\]

---

## 🔹 6. Gaussian Distribution
A symmetrical, bell-shaped distribution where mean, median, and mode are equal.

---

## 🔹 7. 3-Sigma Rule
- 68% within 1σ  
- 95% within 2σ  
- 99.7% within 3σ  

---

## 🔹 8. Percentiles
A measure indicating the value below which a given percentage of observations fall.

---

## 🔹 9. Quartiles
Values that divide data into four equal parts: Q1, Q2 (median), and Q3.

---

## 🔹 10. Five Number Summary
Includes minimum, Q1, median, Q3, and maximum values.

---

## 🔹 11. Skewness
A measure of the asymmetry of a distribution.

---

## 🔹 12. Kurtosis
A measure of the peakedness or flatness of a distribution.

---
